```python
rhoL=1.0
vL=0.0
pL=100000.0
rhoR=0.125
vR=0.0
pR=10000
x0=-10.0
xf=10.0
gamma=1.4
dt=0.00001
N=1000
x=linspace(x0,xf,N)
dx=x[1]-x[0]
```

```python
rho_initial=zeros(N)
v_initial=zeros(N)
p_initial=zeros(N)
for i in range(N):
    if x[i]<0:
        rho_initial[i]=rhoL
        p_initial[i]=pL
    else:
        rho_initial[i]=rhoR
        p_initial[i]=pR
print dt/dx
```

```python
u1_initial=rho_initial
u2_initial=rho_initial*v_initial
u3_initial=(p_initial/(gamma-1)+rho_initial*v_initial*v_initial/2.0)
f1_initial=u2_initial
f2_initial=u2_initial*u2_initial/u1_initial+(gamma-1)*(u3_initial-u2_initial*u2_initial/(2.0*u1_initial))
f3_initial=(u3_initial+(gamma-1)*(u3_initial-u2_initial*u2_initial/(2.0*u1_initial)))*u2_initial/u1_initial
```

```python
#Condiciones de frontera fijas
u1_future=zeros(N)
u2_future=zeros(N)
u3_future=zeros(N)
u1_future[0]=u1_initial[0]
u1_future[N-1]=u1_initial[N-1]
u2_future[0]=u2_initial[0]
u2_future[N-1]=u2_initial[N-1]
u3_future[0]=u3_initial[0]
u3_future[N-1]=u3_initial[N-1]
u1_half=u1_initial
u2_half=u2_initial
u3_half=u3_initial
```

```python
plot(x,p_initial)
title('Presion inicial')
xlabel('x')
ylabel('Presion [Pa]')
grid()
```

```python
plot(x,v_initial)
title('Velocidad inicial')
xlabel('x')
ylabel('Velocidad [m/s]')
grid()
```

```python
plot(x,rho_initial)
title('Densidad inicial')
xlabel('x')
ylabel('Densidad [kg/m^3]')
grid()
```

```python
for i in range(1,N-1):
    u1_half[i]=(u1_initial[i+1]+u1_initial[i])/2.0-(f1_initial[i+1]-f1_initial[i])*dt/(2.0*dx)
    u2_half[i]=(u2_initial[i+1]+u2_initial[i])/2.0-(f2_initial[i+1]-f2_initial[i])*dt/(2.0*dx)
    u3_half[i]=(u3_initial[i+1]+u3_initial[i])/2.0-(f3_initial[i+1]-f3_initial[i])*dt/(2.0*dx)
      
f1_half=u2_half
f2_half=u2_half*u2_half/u1_half+(gamma-1)*(u3_half-u2_half*u2_half/(2.0*u1_half))
f3_half=(u3_half+(gamma-1)*(u3_half-u2_half*u2_half/(2.0*u1_half)))*u2_half/u1_half
    
for i in range(1,N-1):
    u1_future[i]=u1_initial[i]-(f1_half[i]-f1_half[i-1])*dt/(dx)
    u2_future[i]=u2_initial[i]-(f2_half[i]-f2_half[i-1])*dt/(dx)
    u3_future[i]=u3_initial[i]-(f3_half[i]-f3_half[i-1])*dt/(dx)
    
f1_future=u2_future
f2_future=u2_future*u2_future/u1_future+(gamma-1)*(u3_future-u2_future*u2_future/(2.0*u1_future))
f3_future=(u3_future+(gamma-1)*(u3_future-u2_future*u2_future/(2.0*u1_future)))*u2_future/u1_future
```

```python
rho_future=u1_future
v_future=u2_future/rho_future
p_future=(gamma-1)*(u3_future-u2_future*u2_future/(2.0*u1_future))
```

```python
plot(x,v_future)
title('Velocidad en la primera iteracion t=0.00001')
xlabel('x')
ylabel('Velocidad [m/s]')
grid()
```

```python
plot(x,p_future)
title('Presion en la primera iteracion t=0.00001')
xlabel('x')
ylabel('Presion [Pa]')
grid()
```

```python
plot(x,rho_future)
title('Densidad en la primera iteracion t=0.00001')
xlabel('x')
ylabel('Densidad [kg/m^3]')
grid()
```

```python
tf=0.01
Ntime=int(tf/dt)
```

```python
for j in range(1,Ntime):
    for i in range(1,N-1):
        u1_half[i]=(u1_initial[i+1]+u1_initial[i])/2.0-(f1_initial[i+1]-f1_initial[i])*dt/(2.0*dx)
        u2_half[i]=(u2_initial[i+1]+u2_initial[i])/2.0-(f2_initial[i+1]-f2_initial[i])*dt/(2.0*dx)
        u3_half[i]=(u3_initial[i+1]+u3_initial[i])/2.0-(f3_initial[i+1]-f3_initial[i])*dt/(2.0*dx)
          
    f1_half=u2_half
    f2_half=u2_half*u2_half/u1_half+(gamma-1)*(u3_half-u2_half*u2_half/(2.0*u1_half))
    f3_half=(u3_half+(gamma-1)*(u3_half-u2_half*u2_half/(2.0*u1_half)))*u2_half/u1_half
        
    for i in range(1,N-1):
        u1_future[i]=u1_initial[i]-(f1_half[i]-f1_half[i-1])*dt/(dx)
        u2_future[i]=u2_initial[i]-(f2_half[i]-f2_half[i-1])*dt/(dx)
        u3_future[i]=u3_initial[i]-(f3_half[i]-f3_half[i-1])*dt/(dx)
        
    f1_future=u2_future
    f2_future=u2_future*u2_future/u1_future+(gamma-1)*(u3_future-u2_future*u2_future/(2.0*u1_future))
    f3_future=(u3_future+(gamma-1)*(u3_future-u2_future*u2_future/(2.0*u1_future)))*u2_future/u1_future
    
    u1_initial=u1_future.copy()
    u2_initial=u2_future.copy()
    u3_initial=u3_future.copy()
    f1_initial=f1_future.copy()
    f2_initial=f2_future.copy()
    f3_initial=f3_future.copy()
```

```python
rho_future=u1_future
v_future=u2_future/rho_future
p_future=(gamma-1)*(u3_future-u2_future*u2_future/(2.0*u1_future))
```

```python
plot(x,v_future)
title('Velocidad t=0.01')
xlabel('x')
ylabel('Velocidad [m/s]')
grid()
```
