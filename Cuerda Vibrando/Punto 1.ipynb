```python
n_points = 100
L=100.0
x = linspace(0.0,L,n_points)
u_initial=zeros(n_points)
for i in range(n_points):
    if (x[i]<=0.8*L):
        u_initial[i]=1.25*x[i]/L
    else:
        u_initial[i]=5-5*x[i]/L
u=zeros((121,100))
u[0,:]=u_initial
```

```python
plot(x,u[0,:])
```

```python
#find the first iteration for fixed boundary conditions
delta_x = x[1]-x[0]
delta_t = 0.0005
rho=0.001
T=40
c = (T/rho)**0.5
r = c * delta_t / delta_x

print r # remember, this should be less than 1.0 for this scheme to work

#fixed boundary conditions
u_initial[0] = 0.0
u_initial[n_points-1] = 0.0

u_future = zeros(n_points)
u_future[0] = 0.0
u_future[n_points-1] = 0.0

for i in range(1,n_points-1):
    u_future[i] = u_initial[i] + (r**2/2.0) * (u_initial[i+1] - 2.0 * u_initial[i] + u_initial[i-1])

#create a new variable to hold the previous value
u_past = u_initial.copy()
#create a new variable to hold the present value
u_present = u_future.copy()
u[1,:]=u_present
```

```python
plot(x, u[0,:],c='red')
plot(x, u[0,:],c='blue')
plot(x, u[1,:],c='green')
plt.xlabel('x')
plt.ylabel('u')
legend(['Initial','Past','Present'])
```

```python
#next iterations
n_time = 120
t=linspace(0.0,n_time+1,n_time+1)
for j in range(n_time):
    for i in range(1,n_points-1):
        u_future[i] = (2.0*(1.0-r**2))*u_present[i] - u_past[i] + (r**2)*(u_present[i+1] +  u_present[i-1])
    u_past = u_present.copy()
    #create a new variable to hold the present value
    u_present = u_future.copy()
    u[j+1,:]=u_present
```

```python
plot(x, u[0,:])
plot(x, u[-1,:])
plt.xlabel('x')
plt.ylabel('u')
```
